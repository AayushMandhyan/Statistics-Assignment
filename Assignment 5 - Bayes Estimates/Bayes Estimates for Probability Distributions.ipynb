# Assignment 5
Add Bayes estimates and plots of the posterior distribution using conjugate priors for Poisson, Binomial, Normal, exponential, and uniform (0,W)

```r
# For Poisson Distribution
n.val <- 100
x.1 <- rpois(n.val,lambda = 0.5)
poisson_conjugate <- function(vec, alpha, beta) {
  n <- length(vec)
  alpha_new <- alpha + sum(vec)
  beta_new <- beta + n
  x <- seq(-1, 2, length = 10000)
  plot(x, dgamma(x,shape = alpha_new, rate = beta_new))
}
poisson_conjugate(x.1, 4, 2)
```

```r
# For Binomial Distribution
x.2 <- rnbinom(n.val, size = 50, prob = 0.5)
binom_conjugate <- function(vec, alpha, beta, r) {
  n <- length(vec)
  alpha_new <- alpha + (r * n)
  beta_new <- beta + sum(vec)
  x <- seq(0, 1, length = 10000)
  plot(x, dbeta(x, alpha_new, beta_new))
}
binom_conjugate(x.2, 4, 2, 100)
```

```r
# For Normal Distribution
sd.val <- 8
x.3 <- rnorm(n.val, mean = 9, sd = sd.val)
normal_conjugate <- function(vec, mu, sigma, sample_precision, p_precision) {
  n <- length(vec)
  mu.new <- ((mu * p_precision) + (n * sample_precision * mean(vec)))/(p_precision + (n * sample_precision))
  sigma.new <- (p_precision + (n * sample_precision))
  x <- seq(-10, 10, length = 10000)
  plot(x, dnorm(x, mean = mu.new, sd = sqrt(1/sigma.new)))
}
normal_conjugate(x.3, 2, 7, sample_precision = (1/(sd.val^2)), 3)
```

```r
# For Exponential Distribution
x.4 <- rexp(n.val,rate = 0.8)
poisson_conjugate <- function(vec, alpha, beta) {
  n <- length(vec)
  alpha_new <- alpha + n
  beta_new <- beta + sum(vec)
  x <- seq(-1, 2, length = 10000)
  plot(x, dgamma(x,shape = alpha_new, rate = beta_new))
}
poisson_conjugate(x.4, 4, 2)
```

```r
# For Uniform Distribution(0,W)
dpareto <- function(x,alpha,theta){
  (alpha*(theta^alpha)*(x+theta)^(-alpha-1))*(x>=0)
}

x.5 <- runif(n.val, min = 0, max = 10)
uniform_conjugate <- function(vec, alpha, w) {
  n <- length(vec)
  alpha_new <- alpha + n
  w.new <- max(w, max(vec))
  x <- seq(-1, 1, length = 10000)
  plot(x, dpareto(x, alpha_new, w.new))
}
uniform_conjugate(x.5, 4, 3)
```
